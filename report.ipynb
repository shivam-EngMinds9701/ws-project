```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Processed data

```python
df = pd.read_csv('processed_data.csv')
df.head()
```

```python
sns.set_style('whitegrid')
```

### Distribution of prices

```python
sns.histplot(df['price'], bins =10, kde = True)
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price (INR)')
plt.ylabel('Count')
plt.savefig('visuals/price_distribution.png')
plt.show()
```

- The histogram shows that most laptops are priced between 15,000 and 40,000 INR, with a peak in the 15,000 to 25,000 INR range. 

- As the price increases, the number of laptops available decreases, indicating a right-skewed distribution. 

- The density curve superimposed on the histogram highlights this trend, with a long tail extending towards higher prices. 

- Overall, the data suggests that more affordable laptops are more prevalent in the market compared to higher-priced models.

### Most reviewed Laptops Brand


```python
plt.figure(figsize = (12, 8))
reviews_by_brand = df.groupby('brand_name')['review_count'].sum().reset_index()
sns.barplot(x = 'brand_name', y = 'review_count', data = reviews_by_brand)
plt.title('Number of Reviews by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Reviews')
plt.savefig('visuals/review_count_by_brand.png')
plt.show()
```

- The bar chart displays the number of reviews for laptops by brand. 

- HP dominates with nearly 7,000 reviews, far surpassing other brands. ASUS follows with around 2,000 reviews. 

- Other brands, such as Acer, CHUWI, Dell, Infinix, Lenovo, MSI, Mi, Samsung, Ultimus, and Walker, have significantly fewer reviews, all under 500. 

- This suggests HP's strong presence and popularity in the market, as reflected by the high number of customer reviews.

### Best-Rated Laptop Brands 

```python
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'brand_name', y = 'aggregate_rating', data = df)
plt.title('Average Ratings by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.savefig('visuals/avg_rating_by_brand.png')
plt.show()
```

- Rating distribution: Most brands have average ratings between 3.5 and 4.5, with a few exceptions.

- Top performers: MSI stands out with the highest median rating and a large spread, indicating consistently high but variable ratings. Acer and HP also show high median ratings.

- Lower-rated brands: CHUWI and Ultimus have the lowest median ratings, with CHUWI showing a particularly wide spread of ratings.

- Consistency: DELL, Infinix, and Lenovo have relatively compact boxes, suggesting more consistent ratings across their products.

- Outliers: Several brands (e.g., ASUS, Acer, HP) have outlier points, indicating some products that significantly outperform or underperform the brand's typical range.

- Brand variability: ASUS and MSI show large interquartile ranges, suggesting greater variability in product quality or customer satisfaction within these brands.
